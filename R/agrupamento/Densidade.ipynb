<h1>Densidade</h1>

```r
set.seed(42)
folder_name = '.' 
file_name   = 'data2.csv'
```

Leitura dos dados

```r
data = read.csv(file=sprintf('%s/%s',folder_name,file_name),
                sep=',',
                header=TRUE)
```

Análise dos dados

```r
dim(data)
```

```r
data
```

Carrega o pacote DBScan

```r
require(dbscan)
```

Cria o modelo do algoritmo em que:
* x = dados
* eps = raio utilizado para determinar os vizinhos
* minPts = número minimo de pontos para encontrar os pontos 'cores'
* borderPoints = variável booleana que determina se os pontos de bordas devem ou não ser considerados integrantes dos grupos

!!! Atenção !!!<br>
Este algoritmo é extremamente sensível aos parametros eps e minPts... Para melhor compreensão, varie o valor de eps e veja o plot abaixo.

```r
model = dbscan(x = data, 
               eps = 0.03, 
               minPts = 10, 
               borderPoints = TRUE)
```

Calcula o agrupamento

```r
cluster = predict(model, data=data)
```

Número de pontos considerados ruído e, consequentemente, não agrupados

```r
length(which(cluster == 0))
```

Porcentagem de pontos considerados ruído

```r
(length(which(cluster == 0))/nrow(data))*100
```

Como o algoritmo retorna 0 para os pontos que não foi possível agrupar, no plot abaixo os pontos pretos são aqueles que não foram agrupados e, consequentemente, considerados ruídos.

```r
plot(data, col=cluster+1)
```
