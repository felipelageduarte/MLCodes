<h1>Naive Bayes</h1>

```r
library(data.table)
library(MLmetrics)
library(caret)
library(doMC)

set.seed(42)
registerDoMC(cores = detectCores(all.tests = FALSE, logical = TRUE))

folder_name = '.' 
file_name   = 'data.csv'
```

```r
data = fread(file=sprintf('%s/%s',folder_name,file_name))
```

```r
dim(data)
```

```r
data
```

```r
id      = c()
target  = c('Species')
data.vi = names(data)[!(names(data) %in% c(id, target))]
```

```r
f1 <- function(data, lev = NULL, model = NULL) {
  f1_val <- F1_Score(y_pred = data$pred, y_true = data$obs, positive = lev[1])
  c(F1 = f1_val)
}
```

```r
config <- trainControl( 
    #method = "cv", # hold out
    method = "cv",
    #method = "LOOCV", # Leave one out
    number = 10, # number of folds
    
    summaryFunction = f1,
    
    verboseIter = TRUE,
    classProbs  = TRUE,
    allowParallel = TRUE
)
```

```r
model <- train(Species ~ ., 
               data = data,
               method = "nb",
               trControl = config
              )
```

```r
model
```

```r
pred <- predict(model, data, type="raw")
```

```r
pred
```

```r
pred <- predict(model, data, type="raw")
#Warnings that probability is 0 for some cases
confusionMatrix(pred, data$Species)
```
